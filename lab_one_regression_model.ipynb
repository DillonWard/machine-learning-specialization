```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
# input the x and y data
# x is the input feature, y is the target value
x_train = np.array([2.0, 4.0, 6.0])
y_train = np.array([300.0, 600.0, 900.0])
```

```python
m = x_train.shape[0]
print(m)
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
```

```python
# define the type of chart to display
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 150 # 2.0 × 150 = 300.0, 4.0 × 150 = 600.0, 6.0 × 150 = 900.0
b = 0 # base price

def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

plt.plot(x_train, tmp_f_wb, c='b',label='Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# Example: Houses with base price of 50k plus $100 per sqft for construction

x_train = np.array([1.0, 2.0, 3.0, 4.0])
y_train = np.array([150.0, 250.0, 350.0, 450.0])

# Price = 100 * size + 50
w = 100  # 100k per 1000 sqft
b = 50   # 50k base price

predictions = compute_model_output(x_train, w, b)

plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, marker='x', c='r', s=100, label='Actual Values')
plt.plot(x_train, predictions, c='b', linewidth=2, label='Model: Price = 100×Size + 50')

plt.axhline(y=b, color='g', linestyle='--', alpha=0.7, label=f'Base Price = ${b}k')

plt.title("Housing Prices with Base Price")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Example: Multi-City Comparison
x_train = [
    np.array([1.0, 2.0, 3.0, 4.0]), 
    np.array([1.0, 2.0, 3.0, 4.0]), 
    np.array([1.0, 2.0, 3.0, 4.0])
]
y_train = [
    np.array([120.0, 240.0, 360.0, 480.0]),
    np.array([180.0, 360.0, 540.0, 720.0]),
    np.array([130.0, 230.0, 330.0, 430.0])
]
w = [120.0, 180.0, 100.0]
b = [0, 0, 30]
c = ['r', 'b', 'g']
plt.figure(figsize=(10, 6))

for i in range(len(x_train)):
    plt.scatter(x_train[i], y_train[i], marker='x', c=c[i], s=100, label=f'City {i + 1} Data')    
    predictions = compute_model_output(x_train[i], w[i], b[i])
    plt.plot(x_train[i], predictions, c=c[i], linewidth=2, label=f'City {i + 1} Model')

plt.title("Housing Prices for different Cities")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
