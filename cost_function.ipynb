```python
# Cost function formula
the cost function quantifies the error between predicted and actual values
A common cost function for linear regression is the Mean Squared Error (MSE):
`J(w,b) = (1 / 2m) * Σ (f_wb(x_i) - y_i)²`
Example:
x = 5, y = 11, w =2, b = 3, m = 1
w(x)+b = 2(5)+3 = 13
squared_error = (13 - y[i])² = (13 - 11)² = 4
(1/(2 * m)) * squared_error = (1/(2 * 1)) * 4 = 2

## simplified version
fw(x) = wx, b = 0, the calculation is the same but always assume b is 0

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
plt.style.use('./deeplearning.mplstyle')
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

```python
def compute_cost(x, y, w, b=0):
    m = x.shape[0] 
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  
    return total_cost
```

```python
w_range = np.arange(0, 401, 10)
costs = []

for w_val in w_range:
    costs.append(compute_cost(x_train, y_train, w_val))

plt.figure(figsize=(8, 4))
plt.plot(w_range, costs, c='b')
plt.title("Cost vs. w (Simplified Model with b=0)")
plt.xlabel("w (slope)")
plt.ylabel("Cost J(w)")
plt.grid(True)
plt.show()
```

```python
w_vals = np.linspace(0, 400, 100)
b_vals = np.linspace(-100, 300, 100)
J_vals = np.zeros((len(w_vals), len(b_vals)))

for i, w in enumerate(w_vals):
    for j, b in enumerate(b_vals):
        J_vals[i, j] = compute_cost(x_train, y_train, w, b)

min_j_idx = np.unravel_index(np.argmin(J_vals, axis=None), J_vals.shape)
optimal_w = w_vals[min_j_idx[0]]
optimal_b = b_vals[min_j_idx[1]]

costs_for_w = []
for w_val in w_vals:
    costs_for_w.append(compute_cost(x_train, y_train, w_val, optimal_b))

plt.figure(figsize=(8, 4))
plt.plot(w_vals, costs_for_w, c='r')
plt.title(f"Cost vs. w (with b fixed at optimal value of {optimal_b:.1f})")
plt.xlabel("w (slope)")
plt.ylabel("Cost J(w, b_optimal)")
plt.grid(True)
plt.show()
```
